```python
Original source: https://github.com/nadyka/Web-scraping-jobs

Last updated on: 2024-03-20 at 16:20h

# Document the Webscrapers project steps:

# 1. Create a new repository on GitHub by forking the original repository.
# 2. Clone the repository to your local machine.
# 3. Create a new branch for your work.
# 4. Create a new file in the repository.
# 5. Add the file to the staging area.
# 6. Commit the file to the repository.
# 7. Push the file to the repository.

Download the packages:

(base) C:\Windows\System32>cd C:\Users\Nadia Stoyanova

(base) C:\Users\Nadia Stoyanova>conda activate webscrapers

(webscrapers) C:\Users\Nadia Stoyanova> conda install -c conda-forge pandas requests geopy bs4 flask sqlalchemy

###I received the following error: (webscrapers) C:\Users\Nadia Stoyanova>conda install -c conda-forge BeautifulSoup Channels:

conda-forge
defaults Platform: win-64 Collecting package metadata (repodata.json): done Solving environment: failed
PackagesNotFoundError: The following packages are not available from current channels:

beautifulsoup
Current channels:

https://conda.anaconda.org/conda-forge
defaults
To search for alternate channels that may provide the conda package you're looking for, navigate to
https://anaconda.org
and use the search bar at the top of the page.

The error message indicates that the package BeautifulSoup cannot be found in the conda channels you are using. The correct package name for BeautifulSoup in conda is beautifulsoup4, not BeautifulSoup.

You can install it using the following command:
conda install -c conda-forge beautifulsoup4
This command tells conda to install the beautifulsoup4 package from the conda-forge channel.

You can install PyTorch using pip or conda. Here's how you can do it:

Using pip: pip install torch torchvision torchaudio

Using conda: conda install pytorch torchvision torchaudio -c pytorch

These commands will install PyTorch along with torchvision (a package with popular datasets, model architectures, and common image transformations for computer vision) and torchaudio (similar to torchvision, but for audio).

Remember to activate your virtual environment before installing, if you're using one. Also, ensure that your Python and pip versions are up-to-date.
# All requested packages already installed.

(webscrapers) C:\Users\Nadia Stoyanova>conda install pytorch torchvision torchaudio -c pytorch
Channels:
 - pytorch
 - conda-forge
 - defaults
Platform: win-64
Collecting package metadata (repodata.json): done
Solving environment: done

## Package Plan ##

  environment location: C:\ProgramData\Anaconda3\envs\webscrapers

  added / updated specs:
    - pytorch
    - torchaudio
    - torchvision

###Select the fields from indeed.com and from linkedin.com that you want to scrape:
1. Job title
2. Company name
3. Location
4. Zipcode
5. Remote check 
6. In person or hybrid check
6. Salary range
5. Job description: key words, skills, requirements, responsibilities
6. Type of employment: full-time, part-time, contract, internship
7. Benefits
8. Hiring manager name 
9. URL of the job posting
10. Date of job posting to use in a filter
11. Application website
12. Does the applicant require sponsorship?
13. Does the company offer sponsorship?
14. Does the company offer relocation assistance?

### Check if the site has a free/open API:
To find out if a website has an open API (Application Programming Interface), you can follow these steps:

Check the website's documentation: Most websites with an API will have a section in their documentation about it. This is usually found in the footer of the website or in a "Developers" section.

Look for API in the URL: Sometimes, you can find API endpoints by looking at the URLs the website is using. If you see "api" in the URL, it's a good indication that the website has an API.

Use online API directories: Websites like RapidAPI, public-apis.io, and others list thousands of APIs. You can search for the website there.

Contact the website: If you can't find any information about an API, you can try contacting the website directly. They might have an API that isn't publicly advertised.

Remember, even if a website has an API, it might not be open or free to use. Always check the API's terms of use before using it.

APIs, or Application Programming Interfaces, are crucial in software development for several reasons:

Integration: APIs allow different software systems to communicate and interact with each other. This enables developers to integrate functionality from other services into their own applications.

Automation: Through APIs, machines can manage work over the internet. They allow different software programs to communicate with each other and perform tasks automatically.

Efficiency: APIs allow code reuse. Instead of building a functionality from scratch, developers can use APIs to implement features that are already built and tested.

Data Access: APIs often provide a way to access and manipulate data from a service. For example, social media platforms provide APIs for developers to access and post content, or to read user data (with the user's permission).

Expansion: APIs allow a service to be used in more contexts. For example, a web-based service might have a public API so that desktop or mobile apps can also use the service.

Understanding APIs is a fundamental part of modern software development. Whether you're building a web application, a mobile app, or a microservices architecture, you'll likely need to use APIs.
```
